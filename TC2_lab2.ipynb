```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("iris.csv")
```

```python
print(data.describe())
```

```python
data.columns = data.columns.str.strip()
data.hist(bins=20, figsize=(10, 8), color='skyblue', edgecolor='black') 
plt.suptitle("Histograms of Features")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=data)
plt.title("Boxplots of Features")
plt.show()
```

```python

sns.kdeplot(data['sepal_length'], shade=True, color="blue")  # Replace 'sepal_length' with the correct column name if needed

sorted_data = np.sort(data['sepal_length'])  # Replace with the actual column name
cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

plt.figure(figsize=(8, 6))
plt.plot(sorted_data, cdf, marker='.', linestyle='none', color='red')
plt.title("Cumulative Distribution Function (CDF) for Sepal Length")
plt.xlabel('Sepal Length (cm)')
plt.ylabel('CDF')
plt.grid(True)
plt.show()
```

```python
variable = 'sepal_length'

# Ensure the variable name exists in the dataset
count, bins, ignored = plt.hist(data[variable], bins=20, density=True, color='skyblue', edgecolor="black")
bin_centers = 0.5 * (bins[:-1] + bins[1:])  # Calculate the centers of the bins

plt.figure(figsize=(8, 6))
plt.bar(bin_centers, count / sum(count), width=(bins[1] - bins[0]), color='lightgreen', edgecolor='black')

plt.title(f"PMF-like Visualization for {variable.title()}")
plt.xlabel(variable.title())
plt.ylabel('Probability')

plt.show()
```
